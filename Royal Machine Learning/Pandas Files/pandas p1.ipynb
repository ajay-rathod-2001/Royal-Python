```python
series2 = pd.Series([1,2,3,4])
print(series2)
```

```python
empty_s = pd.Series([])
print(empty_s)
```

```python
series3 = pd.Series([1,2,3,4], index=['a','b','c','d',])
print(series3)
```

```python
series3 = pd.Series([1,2,3,4], index=['a','b','c','d',],dtype = float)
print(series3)
```

```python
series3 = pd.Series([1,2,3,4], index=['a','b','c','d',],dtype = float, name ="data values")
print(series3)
```

## Scaler Series :
