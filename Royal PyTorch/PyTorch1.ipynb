 Random Tensors
 
 Why Random Tensors?

 Random tensors are important the way many neural networks learn is that they start with tensors full of random numbers 
 and then adjust those random numbers to bettter represent the data.

 ' Start with random numbers -> look at data -> update random numbers -> look at data -> update random numbers

```python
 # Creating a Random tennsor of size(3,4)
 
rdm_tnsr =torch.rand(3,4)
rdm_tnsr
 
```

Tensor DataTypes :

Note: Tensor datatypes is one of the " 3 Big Errors " you'll run into PyTorch and Deeplearnig
1. Tensors not right datatypes
2. Tensor not right shape
3. Tensor not on the right device

```python
# float 32 tensor

float_32_tensor = torch.tensor([3.0,6.0,9.0],
                               dtype=None,  # what  datatype is the tensor (ex,float32 or float16) 
                               device=None,  # what is your tensor on
                               requires_grad=False) # whether or not to track grandients with this tensors operations
float_32_tensor

```

```python
float_32_tensor.dtype
```

```python
float_16_tensor =float_32_tensor.type(torch.float16)
float_16_tensor
```

```python
float_16_tensor * float_32_tensor
```

```python
int_32_tensor = torch.tensor([3,6,9],dtype=torch.long)
int_32_tensor
```

```python
float_32_tensor * int_32_tensor
```

### Getting information from tensors ( Tensor Attributes)

1. Tensors not right datatypes -to do get datatype from a tensor ,can use 'tensor.dtype'
2. Tensor not right shape - to get shape from a tensor , can use 'tensor.shape'
3. Tensor not on the right device - to get device from a tensor ,can use 'tensor.device'


```python
# Creating a tensor
some_tensor = torch.rand(3,4)
some_tensor
```

```python
some_tensor.size, some_tensor.shape
```

```python
some_tensor.size()
```

```python
# Fing=d out details about some tensor 
print(some_tensor)
print(f"Datatypes of tensor : {some_tensor.dtype} ")
print(f"Shape of tensor : {some_tensor.shape}")
print(f"Device tensor is on : {some_tensor.device}")
```
